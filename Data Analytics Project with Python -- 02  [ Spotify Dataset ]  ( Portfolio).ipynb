```python
# required Libraries

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_t = pd.read_csv("tracks.csv")
```

```python
df_t.head(10)
```

```python
# exploring data
df_t.isnull().sum()
```

```python
df_t.info()
```

```python
# sorting

sort_df = df_t.sort_values("popularity", ascending = False).head(10)
sort_df
```

```python
df_t.describe().transpose()
```

```python
# most popular songs > 90 

most_pop = df_t.query("popularity > 90" , inplace = False).sort_values('popularity' , ascending = False)
most_pop[:10]
```

```python
df_t.set_index("release_date" , inplace = True) # index is set to release date

df_t.index = pd.to_datetime(df_t.index)  # converted time to datetime formate datatupe

df_t.head()
```

```python
# index location
df_t[["artists"]].iloc[18]    # findinf location at perticular no.
```

```python
# changing the duration 

df_t["duration"] = df_t["duration_ms"].apply(lambda x : round(x/1000))

df_t.drop("duration_ms", inplace = True , axis = 1)
```

```python
df_t.duration.head()
```

```python
# Visualization -- 
# 1. heat map

corr_df = df_t.drop(["key","mode","explicit"], axis = 1).corr(method = "pearson")

plt.figure(figsize = (14,6))

heatmap = sns.heatmap(corr_df, annot = True, fmt = ".1g", vmin = -1,vmax = 1, center = 0 , cmap= "inferno", linewidth = 1 , linecolor = "Black")
 # fmt = .1g is string formating for adding annotation
    
heatmap.set_title("Correlation HeatMap Between Variables")

heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
sample_df = df_t.sample(int(0.004*len(df_t)))
```

```python
print(len(sample_df))
```

```python
# scatter plot

plt.figure(figsize = (10,6))

sns.regplot(data = sample_df , y = 'loudness', x = 'energy', color = 'c').set(title = "Loudness vs Energy ")
```

```python
# scatter plot

plt.figure(figsize = (10,6))

sns.regplot(data = sample_df , y = 'popularity', x = 'acousticness', color = 'b').set(title = "Popularity vs Acousticness ")
```

```python
df_t['dates'] = df_t.index.get_level_values('release_date')

df_t.dates = pd.to_datetime(df_t.dates)

years = df_t.dates.dt.year
```

```python
# histogram

sns.displot(years, discrete = True , aspect =2 ,height = 5 , kind = 'hist').set(title = ' Number of songs per year')
```

```python
# pip install --user seaborn == 0.11.0
```

```python
total_dr = df_t.duration
fig_dims = (18,4)

fig,ax = plt.subplots(figsize = fig_dims)

fig = sns.barplot(x = years , y = total_dr,ax=ax, errwidth = False).set(title = "Year vs Duration")
```

```python
# new dataset  genre

df_g = pd.read_csv("SpotifyFeatures.csv")
```

```python
df_g
```

```python
# comapring genre and duration

plt.title("duration odf songs in diffenent genre")
sns.color_palette("rocket" , as_cmap = True)
sns.barplot(y = 'genre', x = 'duration_ms', data = df_g)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
# Top genre on base of popularity

sns.set_style(style = "darkgrid")
plt.figure(figsize = (10,5))
famous = df_g.sort_values("popularity", ascending = False).head(10)
sns.barplot(y = 'genre', x = 'popularity' ,data = famous).set(title = "Top 5 Genres by Popularity")
```
